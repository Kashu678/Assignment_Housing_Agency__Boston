#### Question 3: Provide a boxplot for the MEDV variable vs the AGE variable - Discretize the age variable into three groups of 35 years and younger, between 35 and 50 years and older

Evalution : This question carries 20 marks

```python
# Solution goes here:
bins = [0, 35, 50, boston_df['AGE'].max()]
labels = ['35 years and younger','35 to 50 years','Older than 50 years',]
palette= {'35 years and younger': 'red',
           '35 to 50 years': 'blue',
           'Older than 50 years': 'orange'}
boston_df['AGE'] = pd.cut(boston_df['AGE'],
bins=bins,labels=labels)
sns.boxplot(x='AGE', y='MEDV', palette=palette,data=boston_df)
plt.title('Boxplot of MEDV or AGE variable')
plt.xlabel('Age')
plt.ylabel('MEDV')
plt.show()
```

#### Question 4: Provide a scatter plot to show the relationship between Nitric oxide concentrations and the proportion of non-retail business acres per town. What can you say about the relationship?

Evalution : This question carries 20 marks

```python
# Solution goes here:
plt.scatter(boston_df['INDUS'], boston_df['NOX'], c='red')
plt.xlabel('Proportion of Non-Retail Business Acres per Town')
plt.ylabel('Nitric Oxide Concentrations')
plt.title('Scatter Plot of Nitric Oxide Concentrations or Proportion of Non-Retail Business Acres')
plt.show()


#The Relationship between,Scatter plot shows that  the proportion of non-retail business acres (INDUS) increases, the Nitric Oxide concentrations (NOX) also tend to increase,
# we could infer that areas with more non-retail business activities might be associated with higher pollution levels.
# on the other hand if we talk about negative relation then indus increase or NOX decrease.
```

#### Question 6: Can we conclude that there is no relationship between Nitric oxide concentrations and proportion of non-retail business acres per town? (Pearson Correlation)

Evalution : This question carries 10 marks

```python
# Solution goes here:
correlation_coefficient, p_value = scipy.stats.pearsonr(boston_df['INDUS'], boston_df['NOX'])
print("Pearson correlation coefficient:", correlation_coefficient)
print("p-value:", p_value)
alpha = 0.05
if p_value < alpha:
   print('The correlation is statistically significant')
else:
    print('The correlation is not statistically significant.')
```
